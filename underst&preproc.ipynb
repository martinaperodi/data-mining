```python
def log_transform_columns(df, columns, plot=True):
    """
    Applica una trasformazione logaritmica alle colonne numeriche selezionate.
    Gestisce automaticamente valori 0 o negativi (li sposta di +1 o ignora).

    Parametri:
    -----------
    df : pandas.DataFrame
        Il dataset su cui operare.
    columns : list
        Lista di colonne da trasformare.
    plot : bool
        Se True, mostra la distribuzione prima e dopo la trasformazione.

    Ritorna:
    --------
    df_transformed : pandas.DataFrame
        Una copia del DataFrame con le colonne trasformate (suffisso '_log').
    """
    df_transformed = df.copy()

    for col in columns:
        # Controllo tipo
        if not np.issubdtype(df[col].dtype, np.number):
            print(f" Colonna '{col}' non numerica, salto.")
            continue

        # Evita problemi con 0 o valori negativi
        min_val = df[col].min()
        if min_val <= 0:
            shift = abs(min_val) + 1
            transformed = np.log(df[col] + shift)
            print(f" Colonna '{col}' trasformata con log(x + {shift:.2f})")
        else:
            transformed = np.log(df[col])
            print(f" Colonna '{col}' trasformata con log(x)")

        df_transformed[f"{col}_log"] = transformed

        # Visualizza la distribuzione prima/dopo
        if plot:
            fig, ax = plt.subplots(1, 2, figsize=(10, 4))
            sns.histplot(df[col], bins=40, ax=ax[0], color='lightcoral', kde=True)
            ax[0].set_title(f"{col} - Originale")

            sns.histplot(df_transformed[f"{col}_log"], bins=40, ax=ax[1], color='lightblue', kde=True)
            ax[1].set_title(f"{col}_log - Dopo trasformazione")
            plt.tight_layout()
            plt.show()

    return df_transformed



df_log = log_transform_columns(df2, columns)
```

```python
def detect_outliers_iqr(df, plot=True):
    """
    Rileva outlier in tutte le colonne numeriche del DataFrame usando il metodo IQR.

    Parametri:
    -----------
    df : pandas.DataFrame
        Il dataset da analizzare.
    plot : bool
        Se True, mostra un boxplot per ogni colonna con outlier.

    Ritorna:
    --------
    outlier_report : pandas.DataFrame
        Tabella con numero e percentuale di outlier per colonna.
    """
    results = []

    # Analizza solo le colonne numeriche in df
    numeric_cols = ['YearPublished',
 'GameWeight',
 'MinPlayers',
 'MaxPlayers',
 'ComAgeRec',
 'LanguageEase',
 'BestPlayers',
 'NumOwned',
 'NumWant',
 'NumWeightVotes',
 'ComMinPlaytime',
 'ComMaxPlaytime',
 'MfgAgeRec',
 'NumAlternates',
 'NumExpansions',
 'NumImplementations']

    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        Q4 = df[col].quantile(0.95)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outlier_mask = (df[col] < lower) | (df[col] > upper)
        n_outliers = outlier_mask.sum()
        perc = (n_outliers / len(df)) * 100
        perc95 = (Q4 / len(df)) * 100


        results.append({
            'Colonna': col,
            'Outlier': n_outliers,
            '% Outlier': round(perc, 2),
            'Q1': Q1,
            'Q3': Q3,
            'Q4': Q4,
            'IQR': IQR,
            'Lower Bound': lower,
            'Upper Bound': upper,
            'Perc95': perc95
        })

        if plot and n_outliers > 0:
            plt.figure(figsize=(6, 3))
            sns.boxplot(x=df[col], color='lightblue')
            plt.title(f'Boxplot di {col} ({n_outliers} outlier)')
            plt.tight_layout()
            plt.show()

    # Crea un report ordinato
    outlier_report = pd.DataFrame(results).sort_values(by='Outlier', ascending=False)

    return outlier_report, results

report, results = detect_outliers_iqr(df_log, plot=True)

results
```
